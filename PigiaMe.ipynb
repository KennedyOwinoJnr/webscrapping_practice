## Pagination for the 7 pages

```python
pagesScrapped = 1
df = pd.DataFrame(columns= ['listingID','car_title', 'location', 'car_price', 'currency'] )
landing_page = 'https://www.pigiame.co.ke/car-hire'

url = landing_page

listOfDfs = []

while pagesScrapped < 7:
    # For each page, make a request
    response = requests.get(url=url)

    # Create a soup
    soup = BeautifulSoup(markup=response.content,
                         features='html.parser')
    

    #extact a list of cars

    carlist = soup.findAll(name= 'div', attrs= {'class':'listings-cards__list-item'})

    for idx, car in enumerate(carlist):
    
        listingID = car.find(name='a', attrs= {'class': 'listing-card__inner'})['data-t-listing_id']

        car_title = car.find(name='a', attrs= {'class': 'listing-card__inner'})['data-t-listing_title']

        #grabbing the location
        location = car.find(name= 'a', attrs= {'class':'listing-card__inner'})['data-t-listing_location_title']

        #grabbing the price

        car_price = car.find(name= 'a', attrs= {'class':'listing-card__inner'})['data-t-listing_price']

        #grabbing the listing currency
        currency = car.find(name= 'a', attrs= {'class':'listing-card__inner'})['data-t-listing_currency']

        # Appending information to the dataframe
        df.loc[idx] = [int(listingID),car_title, location, float(car_price), currency]

    # Store the pages info in a list of dataframes
    listOfDfs.append(df)

    # Get the url for the next page
    for i in range(1,8):
        url = f"{landing_page}?page={i}"


    # Update control variable
    pagesScrapped += 1
    
df = pd.concat(listOfDfs, ignore_index= True)
```
